## Examples of activation functions

Step
$$
a(z) = 
\begin{cases}
0, & \text{if } z < 0\\
1, & \text{if } z \geq 0
\end{cases}
$$

Sigmoid
$$
a(z) = \frac{1}{1+\exp{(-z)}}
$$

Hyperboic tangent
$$
a(z) = \tanh(z)
$$

Rectified linear unit (ReLU)
$$
a(z) = \max(0, z)
$$


## Exercise 1: Plot the activation functions specified above over the domain [-5, 5].

```python
import matplotlib.pyplot as plt
import numpy as np

def step(z):
    return np.where(z >=0, 1, 0)

def sigmoid(z):
    return 1/ (1 + np.exp(-z))

def hyperboic_tangent(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)

x = np.linspace(-5, 5, 200)
y_step = step(x)
y_sigmoid = sigmoid(x)
y_tangent = hyperboic_tangent(x)
y_relu = relu(x)

plt.figure(figsize=(10,5))
plt.plot(x, y_step, label='step')
plt.plot(x, y_sigmoid, label='sigmoid')
plt.plot(x, y_tangent, label='hyperboic_tangent')
plt.plot(x, y_relu, label='ReLU')

plt.xlim(-5, 5)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.legend()
```

## Exercise 2: Plot the gradients of the activation functions specified above over the domain [-5, 5].

```python
def derivative(f, z, epsilon=1e-5):
    return (f(z + epsilon) - f(z - epsilon)) / (2 * epsilon)


plt.figure(figsize=(10,5))
plt.plot(x, derivative(step, x), label='step')
plt.plot(x, derivative(sigmoid, x), label='sigmoid')
plt.plot(x, derivative(hyperboic_tangent, x), label='hyperboic_tangent')
plt.plot(x, derivative(relu, x), label='ReLU')

plt.xlim(-5, 5)
plt.ylim(-0.2, 1.2)
plt.grid(True)
plt.legend()
```
